```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, recall_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,VotingClassifier,GradientBoostingClassifier
```

```python
df=pd.read_csv("novagen_dataset.csv")
df.head()
```

```python
#split features and target
x=df.drop("Target",axis=1)
y=df["Target"]
```

```python
y
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42)
```

```python
#Standard Scaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
#LogisticRegression
lr=LogisticRegression(
    penalty="l2",
    solver="liblinear",
    max_iter=1000
)
lr.fit(x_train_scaled,y_train)
y_pred_lr=lr.predict(x_test_scaled)
print("accuracy_score_lr:",accuracy_score(y_test,y_pred_lr))
print("classification_report_lr:",classification_report(y_test,y_pred_lr))
print("recall_score_lr:",recall_score(y_test,y_pred_lr))
```

```python
KNN=KNeighborsClassifier(
    n_neighbors=5,
    metric='minkowski'
)
KNN.fit(x_train_scaled,y_train)
y_pred_knn=KNN.predict(x_test_scaled)
print("accuracy_score_knn:",accuracy_score(y_test,y_pred_knn))
print("classification_report_knn:",classification_report(y_test,y_pred_knn))
print("recall_score_knn:",recall_score(y_test,y_pred_knn))
```

```python
rf=RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42
)
rf.fit(x_train_scaled,y_train)
y_pred_rf=rf.predict(x_test_scaled)
print("accuracy_score_rf:",accuracy_score(y_test,y_pred_rf))
print("classification_report_rf:",classification_report(y_test,y_pred_rf))
print("recall_score_rf:",recall_score(y_test,y_pred_rf))
```

```python
vc=VotingClassifier(
   estimators=[
       ("lr",lr),
       ("rf",rf),
       ("KNN",KNN)
   ],
    voting="soft"
)
vc.fit(x_train_scaled,y_train)
y_pred_vc=vc.predict(x_test_scaled)
print("accuracy_score_vc:",accuracy_score(y_test,y_pred_vc))
print("classification_report_vc:",classification_report(y_test,y_pred_vc))
print("recall_score_vc:",recall_score(y_test,y_pred_vc))
```

```python
gbc=GradientBoostingClassifier(
    n_estimators=150,
    learning_rate=0.1,
    max_depth=4,
    random_state=42
)
gbc.fit(x_train_scaled,y_train)
y_pred_gbc=gbc.predict(x_test_scaled)
print("accuracy_score_gbc:",accuracy_score(y_test,y_pred_gbc))
print("classification_report_gbc:",classification_report(y_test,y_pred_gbc))
print("recall_score_gbc:",recall_score(y_test,y_pred_gbc))
```

we should use RandomForestClassifier for novagen (based on recall) with 94.2 accuracy
